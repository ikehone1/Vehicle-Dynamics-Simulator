```python
#Run 1 Euler formula

m = 10
v0 = 20
vf = 10
d = 50

def Object_Movement(v0, vf, d):
  a = (vf**2 - v0**2) / (2*d) #acceleration
  t_const = d / v0 #constant velocity time
  t_accel = (vf - v0) / a #velocity acceleration time
  return a, t_const, t_accel


a, _, _ = Object_Movement(v0, vf, d)
print(f'Acceleration: {a}')
#print(f'Constant velocity time: {tc}')
#print(f'Accelerating Time: {ta}')

#Euler Integration : v_k+1 = v_k + a*dt or x_k+1 = x_k + v_k*dt
#Euler equation is intergating acceleration to get velocity and integrating velocity to get position
#Position x(t) is the integral of velocity  x(t) = integral v(t) dt
#Velocity v(t) is the derivative of position  v(t) = integral a(t) dt
#Acceleration a(t) is the derivative of velocity  a(t) = dv / dt

dt = 0.01 #timestep in seconds

v = v0
x = 0.0
t = 0.0

time = [] #creates an empty list for time
velocity = []
position = []

while x < d: #v < vf does not work for decreasing velocities
  v = v + a*dt
  x = x + v*dt
  t = t + dt

  time.append(t) #takes the current value of t and adds it to the end of the list. "History of the system" or "Memory"
  velocity.append(v)
  position.append(x)

#Comparing to analytical calculation
t_analytical = (vf - v0) / a

print(f"Euler time: {t:.3f} s") #prints Euler time with 3 decimal places
print(f"Analytical time: {t_analytical:.3f} s")
print(f"Distance error: {x - d:.3f} m")

print(time)
print(velocity)
print(position)

import matplotlib.pyplot as plt
plt.figure()
plt.plot(time, velocity)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Velocity vs Time (Euler Integration)")
plt.grid()
plt.show()

plt.figure()
plt.plot(time, position)
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Position vs Time (Euler Integration)")
plt.grid()
plt.show()
```

```python
#Adding in friction and drag into the simulations while using the Runge-Kutta Method(RK4) instead of Euler's
#I can find both the final velocity at a certain distance and the final distance one velocity hits 0.
import math

m = 10 #kg OBJECT MASS
v0 = 15 #m/s INITIAL VELOCITY
vf = 0.0 #m/s FINAL VELOCITY
d = 0.0 #m DISTANCE
g = 9.81 #m/s^2 GRAVITY
mu = 0.30 #ROAD FRICTION
#Dry Asphalt has a coeff of friction of 0.7-0.9
#Wet Asphalt has a coeff of friction of 0.4-0.6
#Rubber coeff of friction varies from 0.4-1.0
#Steel coeff of fric is 0.5-0.8
Cd = 0.42 #Drag coefficient
#Cube = 1.05
#Sphere = 0.47
#Half Sphere = 0.42
rho = 1.225 #ks/m^3 for air density
A = 1 #m^2 Reference Area

#Adding in Rolling Friction/Resistance          F_net = ma = F_rr  a is the acceleration of the center of mass
C_rr = 0.005 #0.005-0.015  C_rr = ma/mg = a/g   a = alpha*R  alpha is angular acceleration  R is radius
theta = 10 #degrees

#Newtons unit = 1 kg*(m/s**2) FORCE
# Friction Force = u*N
#Normal Force = m*g
N = m*g
print(f'Normal Force: {N} N')
N_angle = m * g * math.cos(theta)

F_friction = N * mu
print(f'Friction Force: {F_friction} N')

def acceleration(v0): #No positive acceleration because no force propelling object forward faster
  a_friction = -mu * g #negative acceleration due to friction
  a_drag = -(rho * Cd * A * v0 * abs(v0)) / (2 * m) #Drag Acceleration uses a = F/mass not F_drag = Cd * A * ((rho * v0**2)/2) Drag Force not Drag Acceleration
  return a_friction + a_drag

a_rr = -C_rr * g * math.cos(theta)
print("Rolling Resistance: {a_rr}")

a = acceleration(v0)
print(f'Acceleration: {a} m/s^2')

#analytical calculations
t_analytical = (vf - v0) / a
d_stop = v0**2 / (2*mu*g)
t_stop = v0 / (mu*g)
print(f"Stopping distance: {d_stop} m")
print(f"Stop time: {t_stop} sec")

dt_e = 0.0001

# Euler and RK4 Computation
def Euler_solver(v0, dt_e):
  dt = dt_e
  v_euler = v0
  t_euler = 0.0
  x_euler = 0.0

  time_e = [] #creates an empty list for time
  velocity_e = []
  position_e = []

  while v_euler > 0:
    a_euler = acceleration(v_euler)
    v_euler += a_euler*dt
    x_euler += v_euler*dt
    t_euler += dt

    time_e.append(t_euler) #takes the current value of t and adds it to the end of the list. "History of the system" or "Memory"
    velocity_e.append(v_euler)
    position_e.append(x_euler)

  import csv
  filename = f"Euler_Friction_Run.csv"
  with open(filename, "w", newline="") as f:
    writer = csv.writer(f)

    #Header
    writer.writerow([
        "Time_s",
        "Velocity_m_per_s",
        "Position_m"
    ])

    #Time History
    for t, v, x in zip(time_e, velocity_e, position_e):
      writer.writerow([t, v, x])

  print(f'Saved {filename}')
  return v_euler, x_euler, t_euler, time_e, velocity_e, position_e

v_euler, x_euler, t_euler, time_e, velocity_e, position_e = Euler_solver(v0, dt_e)

dt_rk4 = 0.0001

def RK4_Solver(v0, dt_rk4):
  dt = dt_rk4
  v_rk4 = v0
  t_rk4 = 0.0
  x_rk4 = 0.0

  print(v_rk4)

  Time = []
  Velocity = []
  Position = []
  EPS = 1e-6
  # dv/dt = a(v)  dx/dt = v

  while v_rk4 > EPS:
    k1_v = acceleration(v_rk4)*dt
    k1_x = v_rk4*dt

    k2_v = acceleration(v_rk4 + 0.5*k1_v)*dt
    k2_x = (v_rk4 + 0.5*k1_x)*dt

    k3_v = acceleration(v_rk4 + 0.5*k2_v)*dt
    k3_x = (v_rk4 + 0.5*k2_x)*dt

    k4_v = acceleration(v_rk4 + k3_v)*dt
    k4_x = (v_rk4 + k3_x)*dt

    v_rk4 += (k1_v + 2*k2_v + 2*k3_v + k4_v)/6
    x_rk4 += (k1_x + 2*k2_x + 2*k3_x + k4_x)/6
    t_rk4 += dt

    Time.append(t_rk4)
    Velocity.append(v_rk4)
    Position.append(x_rk4)

    if v_rk4 < 0:
      v_rk4 = 0
      break

  import csv
  Filename = f"RK4_Friction_Run.csv"
  with open(Filename, "w", newline="") as f:
    writer = csv.writer(f)

    #Header
    writer.writerow([
        "Time_s",
        "Velocity_m_per_s",
        "Position_m"
    ])

    #Time History
    for t, v, x in zip(Time, Velocity, Position):
      writer.writerow([t, v, x])

  print(f'Saved {Filename}')
  return v_rk4, x_rk4, t_rk4, Time, Velocity, Position

v_rk4, x_rk4, t_rk4, Time, Velocity, Position = RK4_Solver(v0, dt_rk4)

print(f'Euler time: {t_euler:.3f} s')
print(f"RK4 time: {t_rk4:.3f} s") #prints RK4 time with 3 decimal places
print(f"Analytical time: {t_analytical:.3f} s")
print(f"Euler distance: {x_euler:.3f} m")
print(f"RK4 distance: {x_rk4:.3f} m")


#print(Time)
#print(Velocity)
#print(Position)

plot_graph(time_e, velocity_e, Time, Velocity, 'Time (s)', 'Velocity (m/s)', 'Euler', 'RK4')
plot_graph(time_e, position_e, Time, Position, 'Time (s)', 'Position (m)', 'Euler', 'RK4')
```

```python
#2D Planar Vehicle (No Suspension) going around a bend
import numpy as np
m = 1000 # kg mass of vehicle
r = 25 #m  Radius of the turn
thetadeg_turn = 90 #degrees
theta_turn = np.deg2rad(thetadeg_turn) #converts turn angle from deg to rad
rho = 1.225 # kg/m^3
Cd = 0.5 # Drag Coefficient
A = 4 # m^2 Object Area
vx = 20 #m/s  Initial X Velocity  or  dx/dt
vy = 0 #m/s   Initial Y Velocity  or  dy/dt
x = 0 #m   Starting x position
y = 0 #m   Starting y position
t = 0 #sec

#Centripetal Acceleration: a_c = v**2/r
#Centripetal Force: F_c = m*(a_c)
def Turning_Velocity(vx, vy):
  v = np.sqrt(vx**2 + vy**2) # m/s  Velocity of Object through the curve or the magnitide of the velocity vector tangent to the curve
  return v

v = Turning_Velocity(vx, vy)
print(f'Initial Turning Velocity: {v} m/s')

#dvx/dt = Fx/m
#dvy/dt = Fy/m
#gravity is ignored due to it being flat ground

#Unit Vectors
tan_vec = np.array([vx,vy]) / v # tangent vector in direction of motion
norm_vec = np.array([-tan_vec[1], tan_vec[0]]) # normal vector perpendicular to motion

def Turning_Acceleration(v):
  a_c = v**2/r #acceleration due to centripetal accceleration
  a_drag = -(rho * Cd * A * v**2) / (2 * m) # negative acceleration due to drag
  a_vec = (-a_drag * tan_vec) + (a_c * norm_vec)
  a_x = a_vec[0]
  a_y = a_vec[1]
  return a_vec

a_vec = Turning_Acceleration(v)

a_c = v**2/r
F_c = m * a_c # Centripetal Force
print(f'Turning Acceleration: {a_c} m/s^2')
print(f'Centripetal Force: {F_c} N')

dt_e = 0.001

def Turning_Euler(dt_e, vx, vy):
  dt = dt_e
  vx_euler = vx
  vy_euler = vy
  v_euler = Turning_Velocity(vx, vy)
  x_euler = 0.0
  y_euler = 0.0
  t_euler = 0.0
  theta = 0.0
 # velocity vector

  time_euler = [] #lists to store data
  vel_euler = []
  posx_euler = []
  posy_euler = []

  while theta < theta_turn:  #stops when car turns to specified angle
    v_euler = Turning_Velocity(vx_euler, vy_euler)

    tan_vec = np.array([vx_euler, vy_euler]) / v_euler
    norm_vec = np.array([-tan_vec[1], tan_vec[0]])

    a_vec = Turning_Acceleration(v_euler)

    vx_euler += a_vec[0] * dt
    vy_euler += a_vec[1] * dt
    x_euler += vx_euler * dt
    y_euler += vy_euler * dt
    t_euler += dt
    theta += (v_euler/r) * dt

    time_euler.append(t_euler)
    vel_euler.append(v_euler)
    posx_euler.append(x_euler)
    posy_euler.append(y_euler)

    if v_euler < 1e-6:
      break

  import csv
  filename = f"Turning_Euler_Drag_Run.csv"
  with open(filename, "w", newline="") as f:
      writer = csv.writer(f)

      #Header
      writer.writerow([
          "Time_s",
          "Velocity_m_per_s",
          "PositionX_m",
          "PositionY_m"
      ])

      #Time History
      for t, v, x, y in zip(time_euler, vel_euler, posx_euler, posy_euler):
        writer.writerow([t, v, x, y])

  print(f'Saved {filename}')
  return v_euler, x_euler, t_euler, time_euler, vel_euler, posx_euler, posy_euler

v_euler, x_euler, t_euler, time_euler, vel_euler, posx_euler, posy_euler = Turning_Euler(dt_e, vx, vy)

import matplotlib.pyplot as plt

plt.figure()
plt.plot(posx_euler, posy_euler)
plt.xlabel('X Position (m)')
plt.ylabel('Y Position (m)')
plt.title('Car Trajectory')
plt.show()

plt.figure()
plt.plot(time_euler, vel_euler)
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Car Velocity')
plt.grid()
plt.show()
```
